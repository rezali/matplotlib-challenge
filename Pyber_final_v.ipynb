```python
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
merge_df = pd.merge(city_df, ride_df, on="city")

# Display the data table for preview
merge_df.head()
```

```python
#all the variables
city_grouped = merge_df.groupby(["type", "city", "driver_count"])
fare_mean = city_grouped["fare"].mean()
ride_count = city_grouped["ride_id"].count()
driver_count_by_city = city_grouped["driver_count"].count()
```

```python
# Calculate Type Percents: % of Total Fares by City Type
type_group = merge_df.groupby('type')
fare_by_type = type_group['fare'].sum()
label = ["Rural", "Suburban", "Urban"]
colors = ["blue", "orange", "green"]
# Build Pie Chart
plt.figure(0)
plt.pie(fare_by_type, autopct="%1.1f%%", labels=label)
plt.title("Percent of Total Fares by City Type")
# Save Figure
plt.savefig("Pie_Fares")
plt.show()
```

```python
# Calculate Type Percents: % of Total rides by City Type
rides_group = merge_df.groupby('type')
rides_by_type = rides_group['ride_id'].count()
label = ["Rural", "Suburban", "Urban"]
colors = ["blue", "orange", "green"]
# Build Pie Chart
plt.figure(1)
plt.pie(rides_by_type, autopct="%1.1f%%", labels=label)
plt.pie
plt.title("Percent of Total Rides by Ride Type")
# Save Figure
plt.savefig("Pie_Rides")
# Show Figure
plt.show() 
```

```python
# Calculate Driver Percents
type_grouped = city_df.groupby(["type"])
driver_count_by_type = type_grouped["driver_count"].sum()
label = ["Rural", "Suburban", "Urban"]
colors = ["blue", "orange", "green"]


# Build Pie Chart
plt.figure(2)
plt.pie(driver_count_by_type, autopct="%1.1f%%", labels=label)
plt.pie
plt.title("Percent of Total Drivers by City Type")
plt.show()
# Save Figure
plt.savefig("Pie_Drivers")
plt.show()
```
